```python
dataset["user_id"] = dataset["user_id"].astype(float)
dataset["durasi_akses"] = dataset["durasi_akses"].astype(float)

print(dataset.info())
```

```python
import numpy as np

# Menambahkan missing value pada kolom tertentu
dataset.loc[10:20, 'jenis_kelamin'] = np.nan
dataset.loc[30:40, 'email'] = np.nan

# Menambahkan outlier pada kolom durasi_akses
dataset.loc[5, 'durasi_akses'] = 1000

# Menambahkan data duplikat
dataset = pd.concat([dataset, dataset.sample(5)])

# Menyimpan ulang indeks setelah penambahan data duplikat
dataset.reset_index(drop=True, inplace=True)
```

```python
# Menghapus baris yang mengandung missing value
dataset = dataset.dropna()

# Tentukan strategi imputasi (misalnya, mengganti nilai hilang dengan median)
imputer = SimpleImputer(strategy="median")

# Pilih kolom dengan nilai numerik untuk diimputasi
numeric_columns = ["user_id", "durasi_akses"]

# Proses imputasi pada kolom-kolom numerik
dataset[numeric_columns] = imputer.fit_transform(dataset[numeric_columns])

# Tampilkan DataFrame setelah imputasi
dataset
```

```python
# Mengatasi data duplikat berdasarkan kolom tertentu
dataset = dataset.drop_duplicates(subset=["user_id", "tgl_akses", "nama_lengkap"])

# Menampilkan DataFrame setelah mengatasi data duplikat
print("\nDataFrame Setelah Mengatasi Data Duplikat:")
print(dataset)
```

```python
#analisis korelasi antara jenis kelamin dan durasi akses

# Menghitung tabel kontingensi
contingency_table = pd.crosstab(dataset["jenis_kelamin"], pd.cut(dataset["durasi_akses"], bins=3, labels=["Rendah", "Sedang", "Tinggi"]))
print("=== Tabel Kontingensi Jenis Kelamin dan Durasi Akses ===")
print(contingency_table)
print("\n")

# Menghitung Cramér's V
chi2, _, _, _ = chi2_contingency(contingency_table)
n = dataset.shape[0]
r, k = contingency_table.shape
cramers_v = np.sqrt(chi2 / (n * min((k-1), (r-1))))
print("=== Hasil Analisis Cramer's V Jenis Kelamin dan Durasi Akses ===")
print(f"Chi-Square: {chi2}")
print(f"Cramer's V : {cramers_v}")
```

```python
# Plotting heatmap untuk matriks kontingensi
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, cmap="Blues", fmt="d", cbar=True)
plt.title("Heatmap Matriks Kontingensi Jenis Kelamin dan Durasi Akses")
plt.xlabel("Durasi Akses")
plt.ylabel("Jenis Kelamin")
plt.show()
```

```python
# Plotting diagram batang untuk distribusi variabel kategorikal pada setiap kategori variabel numerik
plt.figure(figsize=(10, 6))
sns.barplot(x="durasi_akses", y="jenis_kelamin", data=dataset, ci=None, palette="muted")
plt.title("Distribusi Jenis Kelamin pada Setiap Kategori Durasi Akses")
plt.xlabel("Durasi Akses")
plt.ylabel("Jenis Kelamin")
plt.show()
```
